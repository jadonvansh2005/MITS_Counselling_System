```python
df.columns = df.columns.str.strip().str.replace(" ", "")

print("Columns:", df.columns.tolist())
```

```python
missing = df.isnull().sum()
print("\nMissing Values:\n", missing)
```

```python
num_cols = ["OpeningRank", "ClosingRank", "BranchSeats", "TotalSeats", "TotalCandidates"]

for col in num_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python
# Drop only critical rows
df = df.dropna(subset=["OpeningRank", "ClosingRank"])

# Fill remaining
df["BranchSeats"].fillna(df["BranchSeats"].median(), inplace=True)
df["TotalSeats"].fillna(df["TotalSeats"].median(), inplace=True)

# Very important
df["TotalCandidates"].fillna(method="ffill", inplace=True)
```

```python
duplicates = df.duplicated()

print("Duplicate rows:", duplicates.sum())
```

```python
df = df.drop_duplicates()
```

```python
print("Final Shape:", df.shape)
print("\nRemaining Missing:\n", df.isnull().sum())
```

```python
df
```

```python
df["ClosingPercentile"] = 1 - (df["ClosingRank"] / df["TotalCandidates"])
df["ClosingPercentile"] = df["ClosingPercentile"].clip(0, 1)

df["OpeningPercentile"] = 1 - (df["OpeningRank"] / df["TotalCandidates"])
df["OpeningPercentile"] = df["OpeningPercentile"].clip(0, 1)
```

```python
df["BranchShare"] = df["BranchSeats"] / df["TotalSeats"]
```

```python
df["RankRange"] = df["ClosingRank"] - df["OpeningRank"]
df["MidRank"] = (df["OpeningRank"] + df["ClosingRank"]) / 2
```

```python
min_year = df["Year"].min()

df["YearWeight"] = 1 + 0.05 * (df["Year"] - min_year)
```

```python
group_cols = ["Branch", "Category", "Round", "Domicile"]

grouped = df.groupby(group_cols)
```

```python
stats = grouped.agg({

    "ClosingPercentile": [
        "min",
        "max",
        "mean",
        "std"
    ],

    "TotalSeats": "mean",

    "BranchSeats": "mean",

    "TotalCandidates": "mean"
    
}).reset_index()

stats.columns = [

    "Branch",
    "Category",
    "Round",
    "Domicile",

    "MinP",
    "MaxP",
    "MeanP",
    "StdP",

    "TotalSeats",
    "BranchSeats",
    "TotalCandidates"
]
# FIX IMPORTANT
stats["StdP"] = stats["StdP"].fillna(0)

stats.head(50)
```

```python
import numpy as np

# 🔥 Using ClosingPercentile instead of MidPercentile

def calculate_trend(group):

    group = group.sort_values("Year")

    x = group["Year"].values
    y = group["ClosingPercentile"].values   # ✅ CHANGED HERE

    # Remove invalid values
    mask = np.isfinite(x) & np.isfinite(y)

    x = x[mask]
    y = y[mask]

    # Safety checks
    if len(x) < 2:
        return 0.0

    if np.std(x) == 0 or np.std(y) == 0:
        return 0.0

    # Center years for numerical stability
    x = x - x.mean()

    try:
        slope = np.polyfit(x, y, 1)[0]

        # Optional clipping (recommended)
        slope = np.clip(slope, -0.1, 0.1)

        return float(slope)

    except:
        return 0.0
```

```python
group_cols = ["Branch", "Category", "Round", "Domicile"]  # + "Domicile" if you include it

trend_df = (
    df.groupby(group_cols)
      .apply(calculate_trend, include_groups=False)
      .reset_index(name="Trend")
)
```

```python
trend_df["Trend"] = trend_df["Trend"].clip(-0.1, 0.1)
```

```python
final_df = stats.merge(trend_df, on=group_cols)

final_df.to_csv("../data/processed/final_df.csv", index=False)

print("✅ Feature engineering complete")
```

```python
import pandas as pd
import numpy as np

# Raw dataset
raw_df = pd.read_csv("../data/raw/Counselling_system_Dataset.csv")
raw_df.head()
```

```python
# Feature dataset
final_df = pd.read_csv("../data/processed/final_df.csv")
final_df.head()
```

```python
# Show all available years in the dataset
available_years = sorted(raw_df["Year"].unique())

print("Available Years:", available_years)

# Choose any year manually
selected_year = int(input("Enter the year you want to select: "))

# Check if the year exists in dataset
if selected_year in available_years:
    print("Selected Year:", selected_year)

else:
    print("Year not found in dataset.")
```

```python
CURRENT_TOTAL_CANDIDATES = (
    raw_df[raw_df["Year"] == selected_year]["TotalCandidates"]
    .iloc[0]
)

print("Current Total Candidates:", CURRENT_TOTAL_CANDIDATES)
```

```python
user_rank = 91040
user_category = "UR/X/OP"
user_round = "R1"
user_domicile = "Y"
```

```python
CURRENT_TOTAL_CANDIDATES = int(CURRENT_TOTAL_CANDIDATES)

user_percentile = 1 - (user_rank / CURRENT_TOTAL_CANDIDATES)

print("User Percentile:", user_percentile)
```

```python
filtered_df = final_df[
    (final_df["Category"] == user_category) &
    (final_df["Round"] == user_round) &
    (final_df["Domicile"] == user_domicile)
]

filtered_df.head()
```

```python
filtered_df["ExpectedClosingRank"] = (
    (1 - filtered_df["MeanP"]) *
    CURRENT_TOTAL_CANDIDATES
)

filtered_df.head()
```

```python
filtered_df["RankGap"] = (
    filtered_df["ExpectedClosingRank"] -
    user_rank
)

filtered_df.head()
```

```python
def calculate_gap_score(gap):

    if gap >= 120000:
        return 95

    elif gap >= 70000:
        return 85

    elif gap >= 30000:
        return 70

    elif gap >= 10000:
        return 55

    elif gap >= 0:
        return 40

    elif gap >= -10000:
        return 25

    else:
        return 10
```

```python
filtered_df["GapScore"] = (
    filtered_df["RankGap"]
    .apply(calculate_gap_score)
)

filtered_df.head()
```

```python
def calculate_probability(user_p, mean_p, std_p, trend):
    
    # Stability adjustment
    adjusted_mean = mean_p + trend
    
    # Difference
    delta = user_p - adjusted_mean
    
    # Sigmoid
    k = 25
    
    probability = 1 / (1 + np.exp(-k * delta))
    
    # Std penalty
    probability *= (1 - min(std_p, 0.3))
    
    return round(probability * 100, 2)
```

```python
results = []

for _, row in filtered_df.iterrows():

    percentile_prob = calculate_probability(
        user_percentile,
        row["MeanP"],
        row["StdP"],
        row["Trend"]
    )

    gap_score = row["GapScore"]

    # HYBRID FUSION
    final_probability = (
        0.7 * percentile_prob +
        0.3 * gap_score
    )

    results.append({
        "Branch": row["Branch"],
        "ExpectedClosingRank": int(row["ExpectedClosingRank"]),
        "RankGap": int(row["RankGap"]),
        "PercentileProb": round(percentile_prob, 2),
        "GapScore": round(gap_score, 2),
        "FinalProbability": round(final_probability, 2)
    })

results_df = pd.DataFrame(results)

results_df.head(20)
```

```python
def classify(prob):

    if prob >= 85:
        return "Very Safe"

    elif prob >= 70:
        return "Safe"

    elif prob >= 50:
        return "Moderate"

    elif prob >= 30:
        return "Risky"

    else:
        return "Dream"
    
results_df["CategoryType"] = (
    results_df["FinalProbability"]
    .apply(classify)
)
```

```python
results_df = results_df.sort_values(
    by=[
        "FinalProbability"
    ],
    ascending=False
)

results_df.head(20)
```
